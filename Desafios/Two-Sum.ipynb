Dado um array de números nums e um valor target, você deve encontrar os índices de dois números que, somados, resultem no target.Regra: Você não pode usar o mesmo elemento duas vezes.Exemplo: nums = [2, 7, 11], target = 9. Resposta: [0, 1] (porque $2 + 7 = 9$).

```polyglot-notebook
//Resolvendo com forca bruta
public int[] TwoSumBruteForce(int[] nums, int target) {

    for (int i = 0; i < nums.Length; i++)
    {
        for (int j = i + 1; j < nums.Length; j++)
        {
            if (nums[i] + nums[j] == target)
            {
                return new int[] { i, j };
            }
        }
    }

return new int[] { -1, -1 }; 
}
```
