O(1) - Constante: tempo de execução não muda. Ex: Acessar um elemento deum array pelo indice.

O(n) - Linear: O tempo crece proporcionalmente ao numero de itens. procurar um valor em uma lista desordenada usando foreach.

O(n^2) - Quadratica: O tempo crece exponencialmente (o "vilão" da performace)Ex: dois loops for aninhados

O(log n) - Logatitimica: o tempo cresce muito pouco mesmo com muitos dados. ex: busca binria em uma lista ordenada
