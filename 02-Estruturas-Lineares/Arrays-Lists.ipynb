```polyglot-notebook
using System.Diagnostics;

//função para Simular O(log n) com Binarie Search

//void AlgoritimoLogaritmico(int n){
List<int> list = Enumerable.Range(0,10000).ToList();


//add elemento na lista



int posicao = -1;
int n = 7800;
//indexação dos elementos
int fist = 0;
int last = list[list.Count -1];

while (fist <= last)
{

    
    //calcular meio 
    int mid = (fist + last) / 2;


    if (list[mid] == n)
    {

        posicao = mid;
         break;
    } 
    
    if(list[mid] > n){
        last = mid - 1;
    }

    if(list[mid] < n)
    {
        last = mid + 1;
    }

}

//}
//logica com error -> teste 4m e 38s
Console.WriteLine($"o alvo {n} foi encontrado na posição {posicao}")
```
